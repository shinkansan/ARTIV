```python
G = ox.graph_from_file("OSM_study/AutonomouStuff_20191119_134123.osm", simplify=False)
# https://osmnx.readthedocs.io/en/stable/osmnx.html#osmnx.core.graph_from_file

fig, ax = ox.plot_graph(G, node_size=0, edge_linewidth=0.5)
```

```python
nodes, edges = ox.graph_to_gdfs(G)

print("Nodes:\n", nodes.head(), '\n')
print("Edges:\n", edges.head(), '\n')
print("Type:", type(edges))


fig = plt.figure()
ax = fig.add_subplot(1,1,1)

edges.plot(ax=ax, linewidth=1, edgecolor='#BC8F8F')
```

```python
route = nx.shortest_path(G, np.random.choice(G.nodes), 
                         np.random.choice(G.nodes))

ox.plot_graph_route(G, route, fig_height=10, fig_width=10)
```
